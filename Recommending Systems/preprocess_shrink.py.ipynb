```python
from __future__ import print_function, division
from builtins import range, input

import pickle
import numpy as np
import pandas as pd
from collections import Counter

# Veriyi okuyoruz.
# https://www.kaggle.com/grouplens/movielens-20m-dataset
df = pd.read_csv("C:/Users/Halil İbrahim Toprak/Desktop/archive/edited_rating.csv")
print("Orjinal dataframe boyutu:", len(df))

# N kullanıcıların sayısı
N = df.userId.max() + 1
# Filmlerin sayısı
M = df.movie_idx.max() + 1 

user_ids_count = Counter(df.userId)
movie_ids_count = Counter(df.movie_idx)

#Tutmak istediğimiz kullanıcıların ve filmlerin sayısı
n = 10000 # kullanıcı sayısı
m = 2000  # film sayısı
#N kare algoritması olduğu için sayıları düşük aldık.

#En çok geçen film ve kullanıcı idlerini most common metotlarıyla bulabiliyoruz.
#Most common bize tuple döndürüyor.İlk değer anahtar ikinci değer adet.
#Döngüyle beraber sadece en çok geçenlerin id'lerini aldık.
user_ids = [u for u, c in user_ids_count.most_common(n)]   #En çok geçen user idleri
movie_ids = [m for m, c in movie_ids_count.most_common(m)] #En çok geçen film idleri

#Sql kodu ile sadece user id ve movie idlerin kopyalarını alarak veriyi küçülttük.
#Sql kodu select* from dataframe where user_id in user_ids and movie_id in movie_ids
#is in metoduyla içinde değer varsa true döndürüyor yoksa false döndürüyor.
#Bu sayede tuttuğumuz her satır user id ve movie id'ye sahip oluyor.
#copy fonksiyonunu kullandık çünkü aksi taktirde yeni data frame'e aktaramazdık.Copy olmasaydı sadece gösterirdi.
#Hafızada yalnız orjinal data frame yer alır.Diğeri sadece gösterilir değiştirilmez.
#Bir satır seçtiğimiz zaman o satırı gösteren bir işaretçiyle orijinal data frame'in bir alt kümesini göstermiş oluruz.
#Yani overwrite yapmak istiyosak copy fonksiyonu gerekli.
df_small = df[df.userId.isin(user_ids) & df.movie_idx.isin(movie_ids)].copy()

#Data frame'imizi küçültmek için en çok geçen filmleri ve kullanıcı idleriyle yeni bir data frame oluşturduk.Most common ile elde
#ettiğimiz en çok geçen elemanları içeren satırları aldık.



#User idlerin ve movie idlerin alt kümesini seçtik.

#User idler ve movie idler sıralı olmadığından yeniden sıralayıp sözlüğe atıyoruz.
#Bir id sözlüğü oluşturup eski idleri 0'dan başlatıp yeniden atıyoruz.
new_user_id_map = {}

i = 0
for old in user_ids:
  new_user_id_map[old] = i
  i += 1
print("i:", i)

#Aynı şekilde bir movide id sözlüğü oluşturup 0'dan başlayıp sıralayarak yeniden atıyoruz.
new_movie_id_map = {}
j = 0
for old in movie_ids:
  new_movie_id_map[old] = j
  j += 1
print("j:", j)

print("Yeni idleri atıyoruz.")
df_small.loc[:, 'userId'] = df_small.apply(lambda row: new_user_id_map[row.userId], axis=1)
df_small.loc[:, 'movie_idx'] = df_small.apply(lambda row: new_movie_id_map[row.movie_idx], axis=1)
# df_small.drop(columns=['userId', 'movie_idx'])
# df_small.rename(index=str, columns={'new_userId': 'userId', 'new_movie_idx': 'movie_idx'})
print("max user id:", df_small.userId.max())
print("max movie id:", df_small.movie_idx.max())

print("Küçük dataframe boyutu:", len(df_small))
df_small.to_csv("C:/Users/Halil İbrahim Toprak/Desktop/archive/very_small_rating.csv", index=False)
```
